```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

##movie export from movielens full dataset
```

```python
movieData = pd.read_csv("movies.csv")
movieData
```

```python
comparableFeatures = ["genres", "keywords", "cast", "director"]
##some have null values, cannot be computed on
for feature in comparableFeatures:
    movieData[feature] = movieData[feature].fillna('')
    ##will fill in for empty features
movieData
```

```python
combinedFeatures = movieData["genres"] + " " + movieData["keywords"] + " " + movieData["cast"] + " " + movieData["director"]

combinedFeatures
```

## Vectorizing the data and Computing Cosine Similarity

```python
def recommendMovie(movie_id):
    vectorizer = TfidfVectorizer()
    vectorizeCombined = vectorizer.fit_transform(combinedFeatures)
    similarity = cosine_similarity(vectorizeCombined)

    similarityScore = list(enumerate(similarity[movie_id])) ## looking at similarity of index 1
    sortedSimilarity = sorted(similarityScore, key = lambda x:x[1], reverse = True)

    return sortedSimilarity

similarMoviesList = recommendMovie(2) ##call this function, input movieId
```

```python
##printing out similar titles given indexes
i = 1
for movieName in similarMoviesList:
    index = movieName[0]
    title = movieData[movieData.index==index]["title"].values[0]
    if i < 10:
        if i == 1:
            print("Giving recommendations for " + title)
        else:
            print(title)
        i+=1

##below is the recommended movies list generated
```
